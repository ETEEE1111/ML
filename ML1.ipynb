```python
import pandas as pd
import numpy as np
import scipy
import plotly.express as px

from scipy import stats
```

```python
df = pd.read_csv('weather_prediction_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

# Очистка датафрейма

Перевод даты и врмени в удобный формат

```python
df['DATE'] = pd.to_datetime(df['DATE'], format="%Y%m%d")
df.head()
```

# Гистограмма

```python
fig = px.histogram(df, x="BASEL_temp_mean", title='Гистограмма средней температуры')
fig.update_layout(xaxis_title="Температура, в градусах Цельсия", yaxis_title="Количество, шт.")
fig.show()
```

# Круговая диаграмма





Группируем по Summary (описание погоды), некоторые состояния встречаются редко, отнесем их в Other state.

```python
DF_counts = df.groupby('BASEL_cloud_cover').size().reset_index(name='Count')
DF_counts = DF_counts.groupby('BASEL_cloud_cover').sum().reset_index()
DF_counts
```

Создаем pie chart

Круговая диаграмма с легендой, на диаграмме проценты распределения

```python
fig = px.pie(DF_counts, values = 'Count', names = 'BASEL_cloud_cover', title = 'Распределение дней по облачности')
fig.update_layout(legend_title = 'Cloud cover')
fig.show()
```

# График (line chart)


Строим линейный график


---

График показывающий колебания влажности в разные года 

```python
fig = px.line(df, x="DATE", y="BASEL_humidity")
fig.update_layout(xaxis_title = 'Время', yaxis_title = 'Влажность',legend_title = 'Тип осадков')
fig.show()
```

График зависимости температуры от времени года

```python
df_one_y = df[df['DATE'].dt.year == 2000]

fig = px.line(df_one_y, x="DATE", y="BASEL_temp_mean")
fig.update_layout(yaxis_title = 'Средняя температура', xaxis_title = 'Время')
fig.show()
```

# Boxplot

```python
fig = px.box(df,y='BASEL_global_radiation',title='Распределение радиации по квартилям')
fig.show()
```
